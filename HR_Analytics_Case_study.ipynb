## Case Study for Data Cleaning 

```python
#importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing dataset
raw_data = pd.read_csv('E:\Data Science\Case Studies\hr-analytics-case-study\general_data.csv')
```

```python
raw_data.head()
```

```python
#checking the information of the data
raw_data.info()
```

```python
#counting the number of null values ineach column of data
raw_data.isnull().sum()
```

```python
#checking if there is any missing data using heat map 

sns.heatmap(raw_data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#to view datatypes
raw_data.dtypes
```

```python
#replacing missing values with the average values

raw_data['NumCompaniesWorked'].fillna(raw_data['NumCompaniesWorked'].mean(),inplace=True)

raw_data['TotalWorkingYears'].fillna(raw_data['TotalWorkingYears'].mean(),inplace=True)
```

```python
#calculating no of missing values in each column 
raw_data.isnull().sum()
```

```python
#raw_data['NumCompaniesWorked']

#removing unwanted data 
raw_data.drop(['StandardHours','EmployeeCount','EmployeeID'],axis=1,inplace=True)
```

```python
raw_data.dtypes
```

```python
#separating out numerical values 
raw_data_num = raw_data[['Age','DistanceFromHome','Education','JobLevel','MonthlyIncome','NumCompaniesWorked','PercentSalaryHike','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear','YearsAtCompany','YearsSinceLastPromotion','YearsWithCurrManager']].copy()
raw_data_num.head()
```

```python
#separating out categorical values
raw_data_cat = raw_data[['Attrition','BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','Over18']].copy()
raw_data_cat.head()
```

```python
#converting categorical data into dummy variables

raw_data_cat = pd.get_dummies(raw_data_cat)
raw_data_cat.head()
```

```python
#merging the numerical data and the categorical data into one 
raw_data_final = pd.concat([raw_data_num,raw_data_cat],axis=1)
raw_data_final.head()
```

```python
#finding the correlation between the data

raw_data_final.corr()
```
